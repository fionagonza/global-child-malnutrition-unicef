```python
import pandas as pd

# 1. Load the original raw file again
#  IMPORTANT: replace the path string below with whatever you used before when it worked.
# Example if your file is in /content:
# df_raw = pd.read_csv("/content/malnutrition-estimates.csv")
df_raw = pd.read_csv("malnutrition-estimates.csv")  # change this if needed

# 2. Keep only the 8 useful columns
keep = [
    'Country', 'Year', 'Stunting', 'Wasting', 'Underweight', 'Overweight',
    'Income Classification', "U5 Population ('000s)"
]
df = df_raw[keep].copy()

# 3. Drop rows where any of the 4 indicators are missing
df = df.dropna(subset=['Stunting', 'Wasting', 'Underweight', 'Overweight'])

# 4. Map Income Classification codes (0,1,2,3) to text labels
income_map = {
    0: 'High income',
    1: 'Upper middle income',
    2: 'Lower middle income',
    3: 'Low income',
    '0': 'High income',
    '1': 'Upper middle income',
    '2': 'Lower middle income',
    '3': 'Low income'
}

df['Income Classification'] = df['Income Classification'].map(income_map)

# 5. Check results
print(df.shape)
print(df['Income Classification'].unique())
df.head()
```

```python
# Save the cleaned dataset for SQL + Tableau + future use
df.to_csv("malnutrition_clean.csv", index=False)
```
